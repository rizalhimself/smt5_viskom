Contoh perbandingan antara global thresholding dan adaptive tresholding.

```python
#load gambar
gambar2 = cv2.imread("/home/rizalhimself/Documents/Pemrograman/SMT5/VisKom/Praktikum3/ImgSource/img2.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
#global tresholding skala 50
retval, global_thresh1 = cv2.threshold(gambar2, 50, 255, cv2.THRESH_BINARY)
#global thresholding skala 100
retval, gloval_thresh2 = cv2.threshold(gambar2, 100, 255, cv2.THRESH_BINARY)
#adaptive thresholding
adaptive_thresh = cv2.adaptiveThreshold(gambar2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 7)
```

```python
#menampilkan hasil thresholding pada sebuah plot
plt.figure(figsize=[18,15]);
plt.subplot(221);
plt.imshow(gambar2, cmap="gray");
plt.title("gambar asli");
plt.subplot(222);
plt.imshow(global_thresh1, cmap="gray");
plt.title("hasil global tresholding (50)");
plt.subplot(223);
plt.imshow(gloval_thresh2, cmap="gray");
plt.title("hasil global thresholding (100)");
plt.subplot(224);
plt.imshow(adaptive_thresh, cmap="gray");
plt.title("hasil adaptif thresholding");
```

Penggunaan Citra Masking

```python
#load gambar foreground dan backgroud
gambar_foreground = cv2.imread("/home/rizalhimself/Documents/Pemrograman/SMT5/VisKom/Praktikum3/ImgSource/img4.jpg")
gambar_background = cv2.imread("/home/rizalhimself/Documents/Pemrograman/SMT5/VisKom/Praktikum3/ImgSource/img3.jpg")
```

```python
#convert gambar foreground dan backgroud ke rgb
gambar_foreground_rgb = cv2.cvtColor(gambar_foreground, cv2.COLOR_BGR2RGB)
gambar_background_rgb = cv2.cvtColor(gambar_background, cv2.COLOR_BGR2RGB)
```

```python
#pengambilan variabel untuk singkronisasi ukuran
logo_w = gambar_foreground_rgb.shape[0]
logo_h = gambar_foreground_rgb.shape[1]
```

```python
#singkronisasi ukuran pada background dengan ukuran foreground
aspect_ratio = logo_w / gambar_background_rgb.shape[1]
dim = (logo_w, int(gambar_background_rgb.shape[0] * aspect_ratio))
gambar_background_rgb = cv2.resize(gambar_background_rgb, dim, interpolation=cv2.INTER_AREA)
```

```python
#menampilkan background dan foreground
plt.subplot(121);
plt.imshow(gambar_foreground_rgb);
print(gambar_foreground_rgb.shape);
plt.title("gambar foreground");
plt.subplot(122);
plt.imshow(gambar_background_rgb);
print(gambar_background_rgb.shape);
plt.title("gambar background");
```
